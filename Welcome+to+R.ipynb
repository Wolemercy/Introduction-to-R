```r
# has no use because of the categorical nature of the variables
ggplot(data = mpg) + geom_point(mapping = aes(x = class, y = drv))
```

```r
# Aesthetics
# using size for an unordered variable is not advised
# ggplot(data = mpg) + 
#     geom_point(mapping = aes(x = displ, y = hwy, size = class))
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy, colour = class))
```

```r
# Left, transparency
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy, alpha = class))
# Right, shape
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy, shape = class))
```

```r
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy, alpha = class), colour = 'blue')
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy), color = "blue")
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, colour= cty))
```

```r
# Stroke aesthetic
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, stroke = 2))
```

```r
# Add logical argument to aesthetics
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, colour = displ < 2))
```

## Facets

```r
ggplot(data = mpg) +
    geom_point(mapping = aes(x = displ, y = hwy)) + 
# to facet on one variable
    facet_wrap(~ class, nrow = 2)
```

```r
ggplot(data = mpg) +
    geom_point(mapping = aes(x = displ, y = hwy)) + 
# to facet on two variables
    facet_grid(drv ~ cyl)
```

```r
ggplot(data = mpg) +
    geom_point(mapping = aes(x = displ, y = hwy)) + 
# to facet on two variables
    facet_grid(. ~ cyl)
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = drv, y = cyl))
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
# argument orientation is (y-axis ~ x-axis)
  facet_grid(drv ~ .)

ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
# argument orientation is (y-axis ~ x-axis)
  facet_grid(. ~ cyl)
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) + 
  facet_wrap(~ class, nrow = 2)
```
